```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random
# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
options = ["stone", "paper", "scissors"]
# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
games_max=5
# Assign a variable to the number of games a player must win to win.
min_wins=2
# Preferably the value will be based on the number of maximum games
# Define a function that randomly returns one of the 3 options.
HUMAN_WIN=2
MACHINE_WIN=1
TIE=0
display = {HUMAN_WIN:"Human wins", MACHINE_WIN:"Machine wins", TIE:"Tie"}

def machine():
    return (random.choice(options))
def human(): # If it is not stone, paper or scissors keep asking until it is.
    entry = input()
    while entry not in options:
        print("Choose between 'stone', 'paper' or 'scissors': ", options)
        entry = input()
    return entry
def one_match():# Define a function that resolves a combat.
    value=TIE
    human_choix=human()        
    machine_choix=machine()
    print(machine_choix) 
    if human_choix=="scissors":
        if machine_choix=="paper":
            value=HUMAN_WIN
        elif machine_choix== "stone":
            value=MACHINE_WIN
        
    elif human_choix  == "paper":
        if machine_choix=="stone":
            value=HUMAN_WIN
        elif machine_choix =="scissors":
            value=MACHINE_WIN
        
    elif human_choix =="stone":
        if machine_choix  == "paper":
            value=MACHINE_WIN
        elif machine_choix=="scissors":
            value=HUMAN_WIN
    return value # Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
# Define a function that shows the choice of each player and the state of the game
```

```python
def has_win(points):
    
    if points>=min_wins:
        return True
    return False
 

def should_stop(wins_h, wins_m, nombre_de_match):
    if has_win(wins_h):
        return True
    elif has_win(wins_m):
        return True
    elif nombre_de_match>=games_max:
        return True
    else:
        return False
    
   # return has_win(wins_h) or has_win(wins_m) or nombre_de_match >= games_max
```

```python
# This function should be used every time accumulated points are updated
# Create two variables that accumulate the wins of each participant
def matches():
    mpoints=0
    hpoints=0
    number_matches=0
    
    
    while not should_stop(hpoints, mpoints, number_matches):  
        match=one_match()
        number_matches+=1
        #print(match)
        if match==MACHINE_WIN:
            print('machine wins')
            mpoints=mpoints+1
        elif match==HUMAN_WIN:
            print('human wins')
            hpoints=hpoints+1
        else:
            print('tie')
  
    return hpoints, mpoints, number_matches

def play():
    h, m, n= matches()
    
    if h>m:
        print("Human win", "Score", "h",h,"m",m, "no de match", n)
    elif h<m:
        print("Machine wins", "Score", "h",h,"m",m, "no de match", n)
    else:
        print("Tie")
```
